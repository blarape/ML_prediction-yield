```python
# importo librer√≠as
import numpy as np
import pandas as pd
```

```python
# cargo datos
df_balance= pd.read_csv("./data/us-balance-quarterly.csv", sep=";")
df_cashflow = pd.read_csv("./data/us-cashflow-quarterly.csv", sep=";")
df_income = pd.read_csv("./data/us-income-quarterly.csv", sep=";")
```

```python
# merge data
merge_keys = ['SimFinId', 'Fiscal Year', 'Fiscal Period']

df_merged = df_balance.merge(df_cashflow, on=merge_keys, how='inner', suffixes=('_income', '_balance'))
df_merged = df_merged.merge(df_income, on=merge_keys, how='inner', suffixes=('', '_cashflow'))
```

```python
# empiezo a quitarme empresas sin inter√©s
valid_tickers = df_merged['Ticker'].value_counts()
valid_tickers = valid_tickers[valid_tickers >= 8].index.tolist()

df_filtered = df_merged[df_merged['Ticker'].isin(valid_tickers)]
df_filtered
```

```python
# empiezo a quitarme nulos
df_filtered.isnull().mean().sort_values(ascending=False)
```

```python
df_filtered = df_filtered[df_filtered.isnull().mean(axis=1) < 0.2]
df_filtered
```

```python
tickers = df_filtered['Ticker'].dropna().unique().tolist()
tickers
```

```python
tickers = [t for t in tickers if 'delisted' not in t.lower()]
tickers
```
